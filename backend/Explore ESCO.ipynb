```python
import pandas as pd

# Load the full dataset
df = pd.read_csv('data/skills_en.csv')

# 1. Let's look at the shape and columns
print(f"Total Rows: {df.shape[0]}")
print(f"Columns: {df.columns.tolist()}")

# 2. Look at the first 10 rows to see the structure
display(df.head(10))

# 3. THE "PYTHON" INVESTIGATION
# Let's search for any skill that contains the word "python"
python_search = df[df['preferredLabel'].str.contains('python', case=False, na=False)]
print("\n--- Search Results for 'Python' ---")
display(python_search[['preferredLabel', 'description']])

# 4. THE "RECRUITMENT" INVESTIGATION
recruitment_search = df[df['preferredLabel'].str.contains('recruitment', case=False, na=False)]
print("\n--- Search Results for 'Recruitment' ---")
display(recruitment_search[['preferredLabel']].head(10))
```

```python
import re

# 1. CLEANING FUNCTION
def clean_esco_label(label):
    # Remove anything in parentheses: "Python (computer programming)" -> "Python"
    clean = re.sub(r'\(.*\)', '', label)
    return clean.strip().lower()

# 2. EXPLORE ALT LABELS
# Let's see what's inside the 'altLabels' for 'show initiative' or others
print("Example Alt Labels for row 8:")
print(df.iloc[8]['altLabels'])

# 3. BUILD A SMARTER MASTER SET
all_valid_names = set()

for _, row in df.iterrows():
    # Add the cleaned preferred label
    all_valid_names.add(clean_esco_label(row['preferredLabel']))
    
    # Add all alt labels (they are separated by \n in the CSV)
    if pd.notna(row['altLabels']):
        synonyms = row['altLabels'].split('\n')
        for s in synonyms:
            all_valid_names.add(s.strip().lower())

print(f"\n--- REFINED DISCOVERY ---")
print(f"Total searchable skill variations: {len(all_valid_names)}")

# 4. TEST THE BROADER SEARCH
search_terms = ["python", "ml", "recruitment", "hiring", "javascript"]
for term in search_terms:
    found = term.lower() in all_valid_names
    print(f"Is '{term}' now found? -> {found}")

# 5. SEARCH FOR 'RECRUIT' PARTIALS TO SOLVE THE MYSTERY
recruit_matches = [s for s in all_valid_names if 'recruit' in s]
print(f"\nFound {len(recruit_matches)} versions of recruitment-related skills. Examples:")
print(recruit_matches[:10])
```

```python
import re

# 1. Load the data
df = pd.read_csv('data/skills_en.csv')

def get_atomic_skill(label):
    # Remove parentheses and their content
    clean = re.sub(r'\(.*\)', '', str(label))
    clean = clean.strip().lower()
    # Only return if it's 1-3 words long
    if len(clean.split()) <= 3:
        return clean
    return None

# 2. Build the Atomic Gold List
atomic_gold_list = set()
for label in df['preferredLabel']:
    skill = get_atomic_skill(label)
    if skill:
        atomic_gold_list.add(skill)

print(f"--- REFINED ATOMIC LIST ---")
print(f"Total Atomic Skills: {len(atomic_gold_list)}")

# 3. Test the 'Problem' words again
test_words = ["python", "javascript", "machine learning", "recruitment", "hiring"]
for word in test_words:
    print(f"Is '{word}' found? -> {word in atomic_gold_list}")

# 4. If 'recruitment' is still False, let's see why
if "recruitment" not in atomic_gold_list:
    print("\n'Recruitment' not found. Searching for similar labels...")
    similar = [s for s in atomic_gold_list if 'recruit' in s]
    print(f"Closest matches: {similar[:5]}")
```

```python
from nltk.stem import SnowballStemmer
import nltk

# Initialize the stemmer
stemmer = SnowballStemmer("english")

# 1. Create a STEMMED Gold List
stemmed_gold_list = set()
for skill in atomic_gold_list:
    # Stem each word in the skill phrase
    stemmed_phrase = " ".join([stemmer.stem(word) for word in skill.split()])
    stemmed_gold_list.add(stemmed_phrase)

print(f"--- STEMMED VALIDATION ---")
print(f"Total Stemmed Root Skills: {len(stemmed_gold_list)}")

# 2. Function to check if a JD word is a skill
# 1. Update the 'is_it_a_skill' function to use partial matching
def is_it_a_skill_optimized(word, gold_roots):
    """
    Checks if the root of the word is a substring of any 
    valid ESCO skill root.
    """
    # Stem the input word
    word_root = " ".join([stemmer.stem(w) for w in word.lower().split()])
    
    # Partial match: is 'recruit' inside 'recruit employe'?
    # We use a generator expression for efficiency
    for skill_root in gold_roots:
        if word_root in skill_root:
            return True
    return False

# 2. TEST THE FINAL HURDLE AGAIN
test_words = ["python", "recruitment", "hiring", "machine learning"]

print(f"--- OPTIMIZED VALIDATION (Partial Matching) ---")
for word in test_words:
    # We pass the 'stemmed_gold_list' we created in Step 1.7
    is_valid = is_it_a_skill_optimized(word, stemmed_gold_list)
    print(f"Is '{word}' a valid skill? -> {is_valid}")
# 3. TEST THE FINAL HURDLE
test_words = ["python", "recruitment", "hiring", "machine learning"]
for word in test_words:
    print(f"Is '{word}' (root: {stemmer.stem(word)}) a skill? -> {is_it_a_skill(word)}")
```
